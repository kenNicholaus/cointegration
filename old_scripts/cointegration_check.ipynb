```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import coint
from tqdm import tqdm
import pickle
import matplotlib.pyplot as plt
```

```python
read_file = '../../../sp500/prices/adjclose.csv'
with open("../cointegration_tickers.pickle",'rb') as f:
            cointegration=pickle.load(f)
```

```python
symbol = 'AMAT'
df = pd.read_csv(read_file, index_col = 0, parse_dates = True)
```

```python
diff = pd.DataFrame()
window = 30
for i in cointegration[symbol]:
    diff[symbol+'_'+i] = df[symbol] - df[i]
    diff['rm_'+symbol+'_'+i] =diff[symbol+'_'+i].rolling(window=window,center=False).mean()
    diff['diff_'+symbol+'_'+i] = diff[symbol+'_'+i] - diff['rm_'+symbol+'_'+i] 
    diff[[symbol+'_'+i,'rm_'+symbol+'_'+i]][-500:].plot()
    df[[symbol,i]][-500:].plot()
    plt.show()
    print('*'*60)
    
    #df[[symbol,i]].plot()
   
```
