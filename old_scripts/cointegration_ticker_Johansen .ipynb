```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import coint
from statsmodels.tsa.vector_ar.vecm import coint_johansen 
from tqdm import tqdm
import pickle
```

```python
read_file = '../../../sp500/prices/adjclose.csv'
```

```python
df = pd.read_csv(read_file, index_col = 0, parse_dates = True)
```

```python
tickers=df.columns.tolist()
len_tickers = len(tickers)
```

```python
symbols = df.columns.tolist()
cointegration = {}

for symbol in symbols:
    temp = []
    df1=df.dropna(subset=[symbol])
    df1.fillna(0,inplace=True)
    for ticker in tickers[:]:
        if symbol != ticker:
            coint_t, p_value, crit_value = coint(df1[symbol].values, df1[ticker].values)
            if p_value < 0.05:
                temp.append(ticker)
    cointegration[symbol] = temp
    print(symbol, len(cointegration[symbol]))
```

```python
print(len(cointegration))
```

```python
with open("../cointegration_tickers.pickle","wb") as f:
        pickle.dump(cointegration,f)   
```
